##### 5.5-Comando Format ||| 5.6- F-string

```python
faturamento = 2000
custo = 500
lucro = faturamento - custo

#5-com format
print('O faturamento da loja foi {}. O custo da loja foi {}. Assim, o lucro da loja foi {}' .format(faturamento, custo, lucro)) 
#chaves significa que naquele local deve entrar uma variavel. ".format(nomeVariavel)" qual variavel ira entrar no local indicado com {}

#6-com f-string
print(f'O faturamento da loja foi {faturamento} e o lucro de {lucro}') #f = format -> formatar para string
```

##### 6.1-If - Condições

```python
meta = 50000
qtde_vendida = 65300

if qtde_vendida >= meta:
    print ("Batemos a meta de vendas de Iphone, vendemos {} unidades" .format(qtde_vendida))
else:
    print(f"Infelizmente não batemos a meta, vendemos {qtde_vendida} unidades. A meta era de {meta} unidades")
```

```python
meta = 0.05
taxa = 0
rendimento = 0.21

if rendimento > meta:
    if rendimento > 0.20:
        taxa = 0.04
        print ("A taxa foi de {}".format(taxa))
    else:
        taxa = 0.02
        print ("A taxa foi de {}".format(taxa))
else:
    taxa = 0 
    print ("A taxa foi de {}".format(taxa))
```

##### 6.3-Elif

```python
venda_funcionario = float(input ("Quanto o funcionario vendeu?"))
meta = 20000
meta0 = 1 * venda_funcionario
meta3 = 1.03 * venda_funcionario
meta7 = 1.07 * venda_funcionario

if venda_funcionario < meta:
    print (meta0)
elif venda_funcionario < (meta*2):
    print (meta3)
else:
    print(meta7)
```

##### 6.5-And e Or

```python
meta_funcionario = 10000
meta_loja = 250000
venda_funcionario = 9000
venda_loja = 260000

if venda_funcionario >= meta_funcionario and venda_loja >= venda_loja:
    bonus = 0.03 * venda_funcionario
    print("Bonus do funcionário foi de {}" .format(bonus))
else:
    print("alguem nao bateu a meta")
#-------------------------------------------------------------------------------
nota_funcionario = 7
meta_nota = 9

if nota_funcionario >= meta_nota or (venda_funcionario >= meta_funcionario and venda_loja >= venda_loja):
    bonus = 0.03 * venda_funcionario
    print ("Bonus do funcionário foi de {}" .format(bonus))
else: 
    print("Funcionário não ganhou bônus")
```

## 7 - Strings - Texto e a importância no Python

```python
email = "lira@gmail.com"
nome = "João Paulo Lira"

print("1-",len(email))#len mostrar qual o tamanho da string
print("2-",email[4])#mostrar qual o caracter nesta posicao
print("3-",email[:-3])#mostra de tras pra frente
print("4-",email[:3])#mostra de frente pra tras
print("5-",email[5:])#mostra da posicao 5 pra frente
print("6-",email[5:8])#mostra do cinco ao 8

print('7-Tamanho do e-mail ' + str(len(email)) + ' caracteres')
print('8-Primeiro Caractere ' + email[0])
print('9-Ultimo Caractere: ' + email[-1])
print('10-Servidor do email: ' + email[5:10])
```

### Operações com String

str -> transforma valor em string<br>
in e not -> Servem para fazer verificações em string<br>
operador + -> concatenar string<br>
len() -> Calcula o tamanho do texto (quantidade de caracteres)
format e {} -> substitui valores<br>
%s -> substitui textos<br>
%d -> substitui números decimais<br>

```python
faturamento = 1000
custo = 500
lucro = faturamento - custo

#Uso de srt() para concatenar com +
print ('O faturamento da loja foi de: ' + str(faturamento) + " o custo foi de " + str(custo) + " o lucro foi de " + str(lucro))

#Uso de format
print('O faturamento foi de: {0}. O custo foi de {1} e o lucro foi de {2}. Lembrando, o faturamento foi de {0}' .format(faturamento, custo, lucro))

#Uso do %s (string) e %d (numero)
print ('O faturamento foi de: %d. O custo foi de %d e o lucro foi de %d.' % (faturamento,custo, lucro))

print('@' in 'lira@gmail.com')
print('@' in 'lira.gmail.com')
```

```python
email = 'lira@gmail.com'
print('Meu e-mail não é {:^50}, show?'.format(email))

custo = 500
faturamento = 270
lucro = faturamento - custo
print('Faturamento foi {:+} e lucro foi {:+}'.format(faturamento, lucro))

custo = 5000
faturamento = 2700
lucro = faturamento - custo
print('Faturamento foi {:+,} e lucro foi {:+,}'.format(faturamento, lucro))

custo = 500
faturamento = 1300
lucro = faturamento - custo
margem = lucro / faturamento
print('Margem de lucro foi de {:.2%}'.format(margem))

custo = 5000
faturamento = 27000
lucro = faturamento - custo
print('Faturamento foi R${:,.2f} e lucro foi R${:,.2f}'.format(faturamento, lucro))

#transformando no formato brasileiro
lucro_texto = 'R${:_.2f}'.format(lucro)
print(lucro_texto.replace('.', ',').replace('_', '.'))

imposto = 0.15758
preco = 100
valor_imposto = round(preco * imposto, 1)
print('Imposto sobre o preço é de {}'.format(valor_imposto))
```
